```python
## System
import sys
import os
from collections import defaultdict

## Pandas
import pandas as pd
import numpy as np

## Plotting 
import matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
## regional ConSplice O/E scores

o_over_e = []

with open("../data/ConSplice.regional.autosome.50bp.txt") as fh:
   
    header = []
    for i,line in enumerate(fh):
        
       
        if line[0] == "#":  
            header = line.strip().split("\t")
            continue
            
        line_dict = dict(zip(header,line.strip().split("\t")))

        #if float(line_dict["ConSplice_O/E"]) >= -2000 and float(line_dict["ConSplice_O/E"]) <= 2000:
            #o_over_e.append(round(float(line_dict["ConSplice_O/E"]),1))
        o_over_e.append(round(float(line_dict["ConSplice_O/E"]),1))
    
```

```python
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
plt.rcParams['svg.fonttype'] = 'none'

fig, ax = plt.subplots(figsize = (20,10))
sns.set(font='Arial', style = "white")
sns.set_context("paper", font_scale = 2)
sns.despine()

sns.histplot(o_over_e, binwidth = 5)

ax.set_xlim(-2000,2000)
ax.set_yscale("log")

plt.ylabel("Count", fontsize = 50)
plt.xlabel("Observed / Expected", fontsize = 50)
ax.tick_params(axis='x', labelsize= 40 )
ax.tick_params(axis='y', labelsize= 40 )

plt.savefig('../figures/figure1_B.svg')
```
